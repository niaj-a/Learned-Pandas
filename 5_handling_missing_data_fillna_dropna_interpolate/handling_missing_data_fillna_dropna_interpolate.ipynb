Handling Missing Data - fillna, interpolate, dropna

```python
import pandas as pd
df = pd.read_csv("weather_data.csv", parse_dates=['day'])
type(df.day[0])
df
```

```python
df.set_index('day', inplace=True)
df
```

Fillna

```python
#Fill all NaN with one specific value
new_df = df.fillna(0)
new_df
```

```python
#Fill na using column names and dict
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'No Event'
})
new_df
```

Use method to determine how to fill na values

```python
new_df = df.fillna(method='ffill')
new_df
```

```python
new_df = df.fillna(method="bfill")
new_df
```

```python
new_df = df.fillna(method="bfill", axis="columns")
new_df
```

```python
new_df = df.fillna(method="ffill", limit=1)
new_df
```

interpolate

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method="time")
new_df
```

Dropna

```python
new_df = df.dropna() #drop if only one value has NaN
new_df
```

```python
new_df = df.dropna(how='all') #drop rows where all values are NaN
new_df
```

```python
new_df = df.dropna(thresh=1) #allow one non missing value
new_df
```
